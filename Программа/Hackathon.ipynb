### 2. Функционал алгоритма

#### 2.1. Функция getFa предназначена для определения аэродинамической силы, действующей на груз в зависимости от скорости движения.

```python
def getFa(V, Fa_data=Fa):
    """
    Function returns aerodynamical force
    according to the value of velocity
    """
    # Set min and max values of Fa if velocity is less/more than limits
    if V <= min(Fa_data["V"]):
        return min(Fa_data["Fa"])
    if V >= max(Fa_data["V"]):
        return max(Fa_data["Fa"])
    
    # Get index of the closest velocity value
    ind = max(Fa_data[Fa_data["V"]<=V].index)
    
    # Save data at the current and the previous index 
    # to calculate approx. value of Fa
    V1 = Fa_data["V"].iloc[ind-1]
    V2 = Fa_data["V"].iloc[ind]
    Fa1 = Fa_data["Fa"].iloc[ind-1]
    Fa2 = Fa_data["Fa"].iloc[ind]
    
    # If velocity is exatly in dataframe, return according value of Fa
    if V == V2:
        return Fa2
    # If not, return value calculated with linear approximatin
    else:
        return (Fa1+(Fa2-Fa1)*((V-V1)/(V2-V1)))
```

#### 2.2. Функция getW предназначена для определения проекций скорости ветра на ось X и Z.

```python
def getW(height, W_data=W):
    """
    Function returns projections of wind velocity
    according to the height
    """
    # Set min and max values of wind velocity if the height is less/more than limits
    if height <= min(W_data["Y"]):
        return (min(W_data["Wx"]), min(W_data["Wz"]))
    if height >= max(W_data["Y"]):
        return (max(W_data["Wx"]), max(W_data["Wz"]))
    
    # Get index of the closest height value
    ind = max(W_data[W_data["Y"]<=height].index)
    
    # Save data at the current and the previous index 
    # to calculate approx. value of Fa
    Wx1 = W_data["Wx"].iloc[ind-1]
    Wx2 = W_data["Wx"].iloc[ind]
    Wz1 = W_data["Wz"].iloc[ind-1]
    Wz2 = W_data["Wz"].iloc[ind]
    Y1 = W_data["Y"].iloc[ind-1]
    Y2 = W_data["Y"].iloc[ind]
    
    # If height is exatly in dataframe, return according value of Wx and Wz
    if height == Y2:
        return (Wx2, Wz2)
    # If not, return value calculated with linear approximatin
    else:
        Wx = Wx1+(Wx2-Wx1)*((height-Y1)/(Y2-Y1))
        Wz = Wz1+(Wz2-Wz1)*((height-Y1)/(Y2-Y1))
        return (Wx, Wz)
```

#### 2.3. Функция getFlight предназначена для определения траектории движения и значений проекций скорости груза на координатные оси в зависимости от времени, прошедшего с момента сбрасывания груза.

```python
def getFlight(H0, V0, alpha, delta_t, m, x0, z0):
    """
    Function calculates the flight trajectory according to delta_t
    Function accepts:
    initial height, initial velocity, angle(degrees), time step, 
    mass & coordinates of discharge point.
    """
    # Set values if they are beyond the limits
    if alpha < -90:
        alpha = -90
    if alpha > 90:
        alpha = 90
    if delta_t < 0:
        delta_t = 0.01
    if delta_t > 0:
        delta_t = 1
    
    g = 9.81
    y0 = H0
    
    # Calculate projections
    V0x = V0*cos(alpha/180*pi)
    V0y = 0
    V0z = V0*sin(alpha/180*pi)
    
    # Initialize output lists
    out_x = [x0]
    out_y = [y0]
    out_z = [z0]
    
    out_Vx = [V0x]
    out_Vy = [V0y]
    out_Vz = [V0z]
    
    out_t = [0]
    
    # Initialize counter
    j = 0
    
    # Calculate Fa for each step
    while(y0 > 0):
        # Calсulate wind velocity
        Wxz = getW(y0)
        Wx = Wxz[0]
        Wz = Wxz[1]
        
        # Calculate resulting velocity
        V_Wx = V0x - Wx
        V_Wy = V0y
        V_Wz = V0z - Wz
        V_W = sqrt(V_Wx**2 + V_Wy**2 + V_Wz**2)
        
        # Calculate force and coefficients
        FA = getFa(V_W)
        if (V_W == 0):
            ka = 0
        else:
            ka = FA/V_W
    
        FAx = -ka*V_Wx
        FAy = -ka*V_Wy
        FAz = -ka*V_Wz
    
        Fx = FAx
        Fy = FAy-m*g
        Fz = FAz
        
        # Calculate acceleration
        ax = Fx/m
        ay = Fy/m
        az = Fz/m
    
        # Calculate coordinates using delta_t
        x = x0 + V0x*delta_t + ax*delta_t**2/2
        y = y0 + V0y*delta_t + ay*delta_t**2/2
        z = z0 + V0z*delta_t + az*delta_t**2/2
        
        # Add coordinates to output list
        out_x.append(x)
        out_y.append(y)
        out_z.append(z)
        
        # Calculate velocity using delta_t
        Vx = V0x + az*delta_t
        Vy = V0y + ay*delta_t
        Vz = V0z + az*delta_t
        
        # Add velocity projections to output list
        out_Vx.append(Vx)
        out_Vy.append(Vy)
        out_Vz.append(Vz)
    
        # Increment counter
        j += 1
        
        # Calculate current time step
        t = delta_t*j
        out_t.append(t)
        
        # Assign previous values 
        x0 = x
        y0 = y
        z0 = z
    
        V0x = Vx
        V0y = Vy
        V0z = Vz
    
    # Create dictionary
    data = {"Time":out_t, "X":out_x, "Y":out_y, "Z":out_z, "Vx":out_Vx, "Vy":out_Vy, "Vz":out_Vz} 
    
    # Create pandas dataframe
    df = pd.DataFrame(data)
    return(df)
```

#### 2.4. Функция getFlight предназначена для определения оптимального угла сбрасывания груза.

```python
def getAlpha(H0, V0, delta_t, m, alpha_min, alpha_max, delta_alpha):
    """
    Function calculates the optimal angle according to 
    flight-time minimization criteria
    Function accepts:
    initial height, initial velocity, time step, mass
    min and max angle (degrees) & angle step.
    """
    
    # Set min and max time step
    if delta_t < 0:
        delta_t = 0.01
    if delta_t > 1:
        delta_t = 1
  
    t_min_out = [0]
    
    # Create range of angle values  
    alpha_seq = np.arange(alpha_min, alpha_max, delta_alpha) 
    
    # Itereate over the angle range to find the optimal angle value
    for alpha_opt in alpha_seq:
        df_other = getFlight(H0, V0, alpha_opt, delta_t, m, 0, 0)
        t_min = df_other["Time"][df_other.shape[0]-1]
        if alpha_opt == alpha_min:
            t_min_out= [t_min]
        else:
            t_min_out.append(t_min)
    
    # Create dictionary
    data = {"Alpha":alpha_seq, "Time_min":t_min_out}
    
    # Create pandas dataframe
    df_alpha = pd.DataFrame(data)
    return(df_alpha)
```

### 3. Функция нахоэдения оптимального значения, объединяющая представленные выше вспомогательные функции.

#### Функция предназначена для определения всех данных, требующихся по заданию

```python
def getDecision(H0, V0, m):
    """
    Function calculates all the required values
    Function accepts:
    initial height, initial velocity, mass.
    """
    # Set initial values
    alpha_min = -90
    alpha_max =  90
    delta_alpha = 5
    delta_t_1 = 0.1
    delta_t_2 = 0.01
    
    # Set values if they are beyond the limits
    if H0 < 10:
        H0 = 10
    if H0 > 1400:
        H0=1400
    if V0 < 0:
        V0 = 0
    if V0 > 250:
        V0 = 250
    if m < 0:
        m = 0.1
    if m > 10000:
        m = 10000
           
    # First "coarse" step: detect optimal angle in range [-90:90] degrees, 
    # 5 degrees angle step and 0.1 s time step
    d = getAlpha(H0, V0, delta_t_1, m, alpha_min, alpha_max, delta_alpha)
  
    alpha_min = min(d["Alpha"].loc[d["Time_min"] == min(d["Time_min"])]) - delta_alpha
    alpha_max = min(d["Alpha"].loc[d["Time_min"] == min(d["Time_min"])]) + delta_alpha
    delta_alpha = 1

    # Second "fine" step: detect optimal angle in smaller range, 
    # 1 degrees angle step and 0.01 s time step
    d2 = getAlpha(H0, V0, delta_t_2, m, alpha_min, alpha_max, delta_alpha)
 
    alpha_min = min(d["Alpha"].loc[d["Time_min"] == min(d["Time_min"])])
    
    # Detect discharge point
    discharge_p = getFlight(H0, V0, alpha_min, delta_t_2, m, 0, 0)
  
    delta_x = -discharge_p["X"][discharge_p.shape[0]-1]
    delta_z = -discharge_p["Z"][discharge_p.shape[0]-1]
    
    # Calculate final flight trajectory
    trajectory = getFlight(H0, V0, alpha_min, delta_t_2, m, delta_x, delta_z)
    trajectory["Alpha"] = alpha_min
    trajectory.loc[trajectory["Y"] < 0, "Y"] = 0
    
    # Leave two decimals
    trajectory = trajectory.round(2)
    return(trajectory)
```

### 4. Проверка алгоритма

```python
# Enter initial parameters
# Height
H0 = 1400
# Velocity
V0 = 250
# Mass
m = 100

result = getDecision(H0, V0, m)
result.tail()
```
