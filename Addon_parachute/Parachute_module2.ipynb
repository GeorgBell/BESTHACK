```python
#vector visualization
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim3d(0, 10)
ax.set_ylim3d(0, 10)
ax.set_zlim3d(0, 10)
ax.quiver(0, 0, 0, 0, 5, 5, length = 4, normalize = True)
ax.quiver(0, 0, 0, 5, 5, 5, length = 4, normalize = True)
ax.quiver(0, 0, 0, 3, 5, 5, length = 4, normalize = True)
plt.show()
```

```python
#transition matrix
import numpy as np 
import scipy.sparse as sp
from itertools import chain
from collections import Counter

transitions = ['A', 'B', 'B', 'C', 'B', 'A', 'D', 'D', 'A', 'B', 'A', 'D']

def rank(c):
    return ord(c) - ord('A')

T = [rank(c) for c in transitions]


M = [np.array([[1,2,3]]) for _ in range(4)]

for (i,j) in zip(T,T[1:]):
    M[i][j] += 1

for row in M:
    n = sum(row)
    if n > 0:
        row[:] = [f/sum(row) for f in row]

for row in M:
    print(row)
```
